```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
cap = cv2.VideoCapture("malaysia.mp4")
cap.set(cv2.CAP_PROP_POS_MSEC, 2000)
```

```python
_, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
```

```python
resized = cv2.resize(frame,(1164,874))
plt.imshow(resized)
```

```python
print(frame.shape)
print(resized.shape)
```

```python
reshaped = np.reshape(resized,(874,582,6))
plt.imshow(reshaped[:,:,0:3])
```
